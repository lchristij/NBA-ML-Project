```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
csvpath = Path('../Project-03/combined_df.csv')
df = pd.read_csv(csvpath)
```

```python
df.head()
```

```python
df.info('include=all')
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.tail()
```

```python
# dropping columns for model tuning

df = df.drop(columns=['A_H/A',
                      'H_+/-',
                      'A_+/-',
                      'H_PTS',
                      'A_PTS',
                      'A_W/L',
                      'H_FGM',
                      'A_FGM',
                      'H_REB',
                      'A_REB',
                      'H_FTM',
                      'A_FTM',
                      'H_FTA',
                      'A_FTA',
                      'H_TEAM',
                      'A_TEAM',
                      'GAMEDATE.1'])
```

```python
df.info('include=all')
```

```python
df['GAMEDATE'] = pd.to_datetime(df['GAMEDATE'])
```

```python
df.info()
```

```python
df.tail(100)
```

```python
df['H_W/L']=df['H_W/L'].apply(lambda x: 1 if x=="W" else 0)
# converting 'H/A' to 0/1's
df['H_H/A'] = df['H_H/A'].apply(lambda x: 1 if x=="vs" else 0)
```

```python
df.tail(40)
```

```python
# separate non numeric data into separate dataframe
dummy_teamstats_df = df[['H_TEAM1','H_TEAM2','A_TEAM1','A_TEAM2']].copy()
teamstats_df =df.drop(['H_TEAM1','H_TEAM2','A_TEAM1','A_TEAM2'], axis=1)
```

```python
# perform get_dummies on non numeric data
dummy_teamstats_df = pd.get_dummies(dummy_teamstats_df, columns=['H_TEAM1','H_TEAM2','A_TEAM1','A_TEAM2'])
```

```python
dummy_teamstats_df.info('include=all')
```

```python
teamstats_df.info()
```

```python
cmb_teamstats_df = pd.concat([df, dummy_teamstats_df], axis=1)
```

```python
cmb_teamstats_df.shape
```

```python
cmb_teamstats_df.info('include=all')
```

```python
# defining 'W/L' column as feature set
X = cmb_teamstats_df.copy()
X.drop(columns=['H_W/L',
       'H_TEAM1',
       'H_TEAM2',
       'A_TEAM1',
       'A_TEAM2'],axis=1, inplace=True)
X.head(16)
```

```python
#print(X.iloc[:,2:32])
X = X.iloc[:,2:32]
```

```python
X.info()
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)
X = scaler.transform(X)
```
