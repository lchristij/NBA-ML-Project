```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
csvpath = Path('../Project-03/teamGameStats.csv')
df = pd.read_csv(csvpath)
```

```python
df.shape
```

```python
df.head(16)
```

```python
home = df[df['H/A'] == 'vs']
home = home.reset_index()
```

```python
home.head()
```

```python
away = df[df['H/A'] == '@']
away = away.reset_index()
```

```python
away.head()
```

```python
home.columns = ['H_' + str(col) for col in home.columns]
away.columns = ['A_' + str(col) for col in away.columns]
```

```python
home.rename(columns={'H_GAMEDATE':'GAMEDATE'}, inplace=True)
away.rename(columns={'A_GAMEDATE':'GAMEDATE'}, inplace=True)
```

```python
home.head(8)
```

```python
away.head(8)
```

```python
home.sort_values(by=['H_TEAM1', 'GAMEDATE'], inplace=True)
```

```python
away.sort_values(by=['A_TEAM2', 'GAMEDATE'], inplace=True)
```

```python
home.head()
```

```python
away.head()
```

```python
home=home.drop(['H_index'], axis=1)
```

```python
away=away.drop(['A_index'], axis=1)
```

```python
home=home.reset_index()
```

```python
away=away.reset_index()
```

```python
home.head()
```

```python
home=home.drop(['index'], axis=1)
```

```python
away=away.drop(['index'], axis=1)
```

```python
home.head()
```

```python
combined_df = pd.concat([home, away], axis=1)
```

```python
combined_df.head()
```

```python
combined_df.info()
```

```python
combined_df['GAMEDATE'] = combined_df['GAMEDATE'].astype(str).apply(pd.to_datetime)
```

```python
combined_df
```

```python
combined_df.shape
```

```python
combined_df.info('include=all')
```

```python
combined_df.to_csv('combined_df.csv')
```
